5. Perform meaningful visualizations using the dataset. Bring at least 3
visualizations.

```python
# 1. Bar Chart of Average Sepal Length by Species

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the iris dataset
# If not already loaded, you would read it like this:
# df = pd.read_excel('iris.xls')

# Calculate average sepal length for each species
avg_sepal_length = df.groupby('Classification')['SL'].mean().reset_index()

# Create a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Classification', y='SL', data=avg_sepal_length)
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length')
plt.show()
```

```python
# 2. Histogram of Petal Length Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['PL'], bins=20, kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Scatter Plot of Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SL', y='SW', hue='Classification', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.show()
```
