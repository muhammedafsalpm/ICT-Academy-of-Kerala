**Theoretical Probability:**
Theoretical probability is based on the assumption that all outcomes of a random experiment are equally likely. It is calculated without conducting any experiments but rather by analyzing the possible outcomes.

Probability of an event occurring can be calculated as:

**Experimental Probability:**

Experimental probability is determined by conducting experiments or observations and recording the outcomes. It is calculated by dividing the number of times an event occurs by the total number of trials.

Suppose you roll a die 100 times and record the outcomes. If you roll a 3 on 20 of those trials, the experimental probability of rolling a 3 is calculated as:

image.png

Eventually, if you roll the die a very large number of times, the experimental probability of rolling each number (1 through 6) will approach 1/6. The larger the sample size, the closer the experimental results will be to the expected theoretical values.

```python
from random import randint
from collections import Counter

# A function to generate a random experiment of rolling a die
# x: number of times the die is rolled
def gen(x):
    exp_results = []  # List to store the results of each die roll
    for i in range(x):
        exp_results.append(randint(1, 6))  # Append a random number between 1 and 6 to the results list
    return exp_results

# Conducting an experiment with 100 rolls of the die
event_first = dict(Counter(gen(100)))  # Count the frequency of each outcome (1 through 6) and convert to dictionary
print("Experiment 1: 100 Rolls")
print("Results of the first experiment with 100 rolls:")
for k, v in event_first.items():
    print("Probability of rolling a", k, "is", v/100)  # Calculate and print the experimental probability for each outcome
print("------------------------------")

# Conducting an experiment with 10,000 rolls of the die
event_second = dict(Counter(gen(10000)))  # Count the frequency of each outcome (1 through 6) and convert to dictionary
print("Experiment 2: 10,000 Rolls")
print("Results of the second experiment with 10,000 rolls:")
for k, v in event_second.items():
    print("Probability of rolling a", k, "is", v/10000)  # Calculate and print the experimental probability for each outcome
```

In summary, theoretical probability provides a model of what we expect in an ideal situation, while experimental probability gives us the actual observed outcomes. Over many trials, the experimental probabilities tend to approximate the theoretical probabilities, illustrating the practical application of probability theory.

**Continuous Random Variables:** These are variables that take on an infinite number of possible values within a given range. *Using the discrete random variables from the dataset we can calculate the probability mass function and using the continuous random variable we can calculate the probability density function.*


---


The **probability density function (PDF)** or **probability mass function (PMF)** determines the probability of a random variable taking a specific value or falling within a particular interval. For discrete distributions, the PMF gives the probability of each possible value, while for continuous distributions, the PDF provides the likelihood of values within a range.

**Probability Mass Function (PMF):** This function applies to discrete random variables and gives the probability that a discrete random variable is exactly equal to some value. It is denoted as P(X = x) for a discrete random variable X taking on a value x. The PMF sums up to 1 over all possible values of the discrete random variable.

**Probability Density Function (PDF):** This function applies to continuous random variables and describes the likelihood of the random variable taking on a particular value within a given range. Unlike the PMF, the PDF does not give probabilities directly; instead, the area under the PDF curve over an interval gives the probability that the continuous random variable falls within that interval. The total area under the PDF curve is equal to 1.

# **Types of Probability Distribution with Examples:**

**1. Discrete Distributions:**

image.png


* **Bernoulli Distribution:** Models a single binary outcome with two possible values (e.g., success or failure, heads or tails).
* **Binomial Distribution:** Represents the number of successes in a fixed number of independent Bernoulli trials (e.g., the number of heads in 10 coin flips).
* **Poisson Distribution:** Describes the number of events that occur in a fixed interval of time or space, assuming a constant rate of occurrence (e.g., the number of emails received per hour).


**2. Continuous Distributions:**

image.png

* **Uniform Distribution:** Provides equal probability to all values within a specified range (e.g., a random number between 0 and 1).
* **Normal Distribution:** Often referred to as the bell curve, it is characterized by a symmetric shape and is widely used in statistical analysis (e.g., heights or weights of a population).
* **Exponential Distribution:** Models the time between consecutive events in a Poisson process (e.g., the time between phone calls at a call center).
* **Gamma Distribution:** Generalizes the exponential distribution and is commonly used to model wait times, failure rates, and various other continuous positive variables.
* **Beta Distribution:** Represents probabilities of events occurring within a fixed interval and is often used as a prior distribution in Bayesian inference.
* **Log-Normal Distribution:** Describes variables that are the product of many small independent factors (e.g., stock prices or incomes).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate class test scores based on a normal distribution
np.random.seed(0)  # For reproducibility
mu, sigma = 75, 10  # Mean test score is 75, standard deviation is 10
test_scores = np.random.normal(mu, sigma, 100)  # Generate 100 test scores

# Plot histogram of the test scores
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(test_scores, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)  #(10,20,100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2)

# Titles and labels
plt.title('Distribution of Class Test Scores')
plt.xlabel('Test Score')
plt.ylabel('Probability Density')

# Show the plot
plt.grid(True)
plt.show()
```

# **Binomial Distribution**

* works on theÂ discrete random variables

* sum of outcomes of an event following a Bernoulli distribution.
* Bernoulli distribution is the simplest discrete distribution, where there are only two possible outcomes:
	* Success (with probability ð‘)
	* Failure (with probability 1âˆ’ð‘)
* Binomial distribution is essentially the sum of outcomes of several Bernoulli trials (repeated experiments). Each trial is independent and has the same probability of success.

* Parameters of Binomial Distribution:
	* **Number of Trials (n)**: The number of times the experiment is conducted. For example, if you toss a coin 10 times, ð‘› = 10.
	* **Probability of Success (p)**: The probability of achieving a success on a single trial. For a fair coin, ð‘=0.5

  Let's consider the example of tossing a fair coin (probability
ð‘ = 0.5 for heads) 10 times ( ð‘›=10):

  * Success: Getting heads
  * Failure: Getting tails

We want to find the probability of getting exactly 3 heads in 10 tosses.



image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # number of trials
p = 0.5  # probability of success

# Binomial distribution
x = np.arange(0, n+1)
binom_pmf = binom.pmf(x, n, p)

# Plotting the distribution
plt.figure(figsize=(10, 6))
plt.bar(x, binom_pmf, color='blue', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.grid(True)
plt.show()
```

## What is poisson distribution?

The Poisson distribution is a probability distribution that expresses the likelihood of a given number of events occurring in a fixed interval of time or space. It is often used to model the number of events that occur within a specific time or space interval, assuming these events happen independently and at a constant average rate.

## Where to Use Poisson Distribution?
* Count Data: It's used when you're interested in counting how many times something happens within a certain period or area.
* Rare Events: It's suitable for events that are rare but can happen multiple times.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
rate = 3  # Average rate (lambda)
# Generate a Poisson distribution
rv = poisson(rate)
x = np.arange(0, 10)  # Values to calculate probabilities for
# Calculate probabilities
pmf = rv.pmf(x)

# Plot the distribution
plt.bar(x, pmf, align='center', alpha=0.5)
plt.xticks(x)
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (Î»={})'.format(rate))
plt.show()
```

2. **Density plot**:
A density plot is similar to a histogram, but instead of showing the number of observations in each bin, it shows the probability density of the data. You can use the kde function from the Seaborn library to plot a density plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.normal(size=1000)

# Plot a density plot
sns.kdeplot(data)

# Add labels to the plot
plt.xlabel('Data Values')
plt.ylabel('Density')

plt.show()
```
